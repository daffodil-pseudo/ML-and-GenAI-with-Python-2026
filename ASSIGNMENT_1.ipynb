```python
# TO FIND AREA OF RECTANGLE
length = float(input("ENTER LENGTH-"))
width = float(input("ENTER WIDTH-"))
area = length*width
print("THE AREA OF RECTANGLE=", area)
```

```python
# TO CALCULATE SIMPLE INTEREST
principal = float(input("ENTER PRINCIPAL AMOUNT="))
rate = float(input("ENTER THE RATE OF INTEREST="))
time = float(input("ENTER THE TIME IN YEARS="))
simple_interest = (principal*rate*time)/100
print("SIMPLE INTERST=", simple_interest)
```

```python
#CONVERSION FROM CELCIUS TO FAHRENHEIT
celcius = float(input("ENTER CELCIUS TEMPERATURE="))
fahrenheit = (celcius*9/5)+32
print("temperature in fahrenheit = ", fahrenheit)
```

```python
# TO CALCULAATE AVERAGE OF THREE NUMBERS
num_1 = float(input("enter the first number = "))
num_2 = float(input("enter the second number = "))
num_3 = float(input("enter the third number = "))
average = (num_1 + num_2 + num_3)/3
print("AVERAGE OF THREE NUMBER = ", average)
```

```python
# TO FIND SQUARE AND CUBE
number = int(input("ENTER A NUMBER = "))
square = number ** 2
cube = number ** 3
print("SQUARE = ", square)
print("CUBE = ", cube)
```

```python
# SWAPPING TWO NUMBERS WITHOUT THIRD VARIABLE
a = int(input("ENTER FIRST NUMBER = "))
b = int(input("ENTER SECOND NUMBER = "))
a,b = b,a
print("AFTER SWAPPING = ")
print("a =",a)
print("b =",b)
```

```python
# TO TAKE STUDENT DETAILS
name = input("ENTER STUDENT NAME = ")
roll_no = input("ENTER ROLL NO = ")
# MARKS DETAIL
maths = float(input("ENTER MATHS MARKS = "))
science = float(input("ENTER SCIENCE MARKS = "))
hindi = float(input("ENTER HINDI MARKS = "))
total_marks= maths+science+hindi
marks_percentage= total_marks/3
#SHOWING RESULTS
print("STUDENT REPORT")
print("NAME OF THE STUDENT = ",name)
print("ROLL NO = ", roll_no)
print("TOTAL MARKS = ", total_marks)
print("PERCENTAGE = ", marks_percentage,"%")
```
